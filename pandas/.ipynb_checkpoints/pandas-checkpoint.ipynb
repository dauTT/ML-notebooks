### [pipe](#index)

<a href="https://pandas.pydata.org/pandas-docs/version/0.21/basics.html#tablewise-function-application">reference</a>



```python
import pandas as pd
# Create empty dataframe
df = pd.DataFrame()

# Create a column
df['name'] = ['John', 'Steve', 'Sarah']
df['gender'] = ['Male', 'Male', 'Female']
df['age'] = [31, 32, 19]

# View dataframe
print(df)

# Create a function that
def mean_age_by_group(dataframe, col):
    # groups the data by a column and returns the mean age per group
    return dataframe.groupby(col).mean()

# Create a function that
def uppercase_column_name(dataframe):
    # Capitalizes all the column headers
    dataframe.columns = dataframe.columns.str.upper()
    # And returns them
    return dataframe

print(df.pipe(mean_age_by_group, col='gender')
         .pipe(uppercase_column_name))
```

<a id='apply'></a>
### [apply](#index)



```python
import numpy as np
df = pd.DataFrame({'one' : pd.Series([1,2,3], index=['a', 'b', 'c']),
                    'two' : pd.Series([1,2,3,4], index=['a', 'b', 'c', 'd']),
                    'three' : pd.Series([5,6,7], index=['b', 'c', 'd'])})


print("df\n",df)

print("\ndf.apply(np.mean)\n\n",df.apply(np.mean))
print("\ndf.apply('mean')\n\n",df.apply('mean'))
print("\ndf.apply(np.mean, axis=1)\n\n",df.apply(np.mean, axis=1))
print("\ndf.apply(lambda x: x.max() - x.min())\n\n",df.apply(lambda x: x.max() - x.min()))

print("\ndf.apply(np.exp)\n\n",df.apply(np.exp))
print("\ndf.apply(lambda x: x*x)\n\n",df.apply(lambda x: x*x))



```

<a id='agg'></a>
### [agg](#index)


```python
import numpy as np
df = pd.DataFrame({'one' : pd.Series([1,2,3], index=['a', 'b', 'c']),
                    'two' : pd.Series([1,2,3,4], index=['a', 'b', 'c', 'd']),
                    'three' : pd.Series([5,6,7], index=['b', 'c', 'd'])})

print(df)
def custom_sum(df):
    return df.sum(skipna=True)

def custom_mean(df):
    return df.mean(skipna=True)

print("\ndf.agg(sum)\n", df.agg(sum))
print("\ndf.agg('sum')\n", df.agg('sum'))
print("\ndf.agg(custom_sum) \n", df.agg(custom_sum))

print("\nspecial agreggation \n", df.agg({'one': ['mean', 'min'], 'two': 'sum', 'three': custom_mean}))

```

<a id='transform'></a>
### [transform](#index)



[example](http://pbpython.com/pandas_transform.html)

Remark. it is similar to the windows function in sql


```python
import pandas as pd
pd.set_option('expand_frame_repr', False)
df=pd.read_excel('sales_transactions.xlsx')
print(df)

#First Approach - Merging
order_total = df.groupby('order')["ext price"].sum().rename("Order_Total").reset_index()
df_1 = df.merge(order_total)
df_1["Percent_of_Order"] = df_1["ext price"] / df_1["Order_Total"]
print("\ndf_1\n", df_1)

#Second Approach - Using Transform

print("aggregation with transform \n",df.groupby('order')["ext price"].transform('sum'))

df["Order_Total"] = df.groupby('order')["ext price"].transform('sum')
df["Percent_of_Order"] = df["ext price"] / df["Order_Total"]
print("\ndf\n", df)

#solution in one line
#df["Percent_of_Order"] = df["ext price"] / df.groupby('order')["ext price"].transform('sum')
```

 ### [applymap](#index) 
 
Since not all functions can be vectorized (accept NumPy arrays and return another array or value), the methods <font color='red'>applymap()</font> on DataFrame and analogously <font color='red'>map()</font> on Series accept any Python function taking a single value and returning a single value
 

```python
import pandas as pd
df=pd.DataFrame(np.arange(15).reshape(5,3))
def f(x):
    return x*x

df[0].applymap(f)
```

### [dropna](#index)

```python
df = pd.DataFrame(np.random.randn(10,3))

df.iloc[::2,0] = np.nan; df.iloc[::4,1] = np.nan; df.iloc[::3,2] = np.nan;

print(df)

#drop all rows that have any NaN values
print('\ndrop all rows that have any NaN values\n',df.dropna())

#drop only if ALL columns are NaN
print('\ndrop only if ALL columns are NaN\n',df.dropna(how='all'))

#Drop row if it does not have at least two values that are **not** NaN
print('\nDrop row if it does not have at least two values that are **not** NaN\n',df.dropna(thresh=2))

#Drop only if NaN in specific column (as asked in the question)
print('\nDrop only if NaN in specific column (as asked in the question)\n',df.dropna(subset=[1]))
```

### [rjust()](#index)

```python
import pandas as pd
df = pd.DataFrame({"Family":['Antares','Delta','Atlas','Falcon', 'xx'],
                    "Launches":[1,1,4,11, 111]})

max_nr_digits=df['Launches'].astype(str).apply(len).max()
df['Family'] =  df['Family']  + '  ' + df['Launches'].astype(str).str.rjust(max_nr_digits) 

print(df)
```

## [recursion](#index)


```python
import pandas as pd
data={'user1':['A','A','B', 'C', 'E', 'F'],
      'user2':['B', 'C','C','D','F','G']}
df=pd.DataFrame(data)

print(df)

# this method is similar to the commnon table expression (CTE) in SQL                                                                    
def cte(df_anchor,df_ref,level):
    if (level==0):
        df_anchor.insert(0, 'user_root',df_anchor['user1'])
        df_anchor['level']=0
        df_anchor['relationship']=df_anchor['user1']+'-'+df_anchor['user2']
        _df_anchor=df_anchor
    if (level>0):
        _df_anchor=df_anchor[df_anchor.level==level]
    _df=pd.merge(_df_anchor, df_ref , left_on='user2', right_on='user1', how='inner', suffixes=('', '_x'))
    
    if not(_df.empty):
        _df['relationship']=_df['relationship']+'-'+_df['user2_x']
        _df['level']=_df['level']+1
        _df=_df[['user_root','user1_x', 'user2_x', 'level','relationship']].rename(columns={'user1_x': 'user1', 'user2_x': 'user2'})
        df_anchor_new=pd.concat([df_anchor, _df])
        return cte(df_anchor_new, df_ref, level+1)
    else:
        return df_anchor
    
df_rel=cte(df, df, 0)
print("\nall relationship=\n",df_rel)

print("\nall relationship related to A=\n", df_rel[df_rel.user_root=='A'])
```
